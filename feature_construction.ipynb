# **Applying Feature Construction**

```python
df.head()
```

```python
df['family_size'] =df['sibsp'] + df['parch'] + 1
```

```python
df.head()
```

```python
def myfun(num):
  if num==1:
    #alone
    return 0
  elif num>1and num<=4:
    #small family
    return 1
  else:
    #large family
    return 2
```

```python
myfun(4)
```

```python
df['family_size']=x['family_size'].apply(myfun)
```

```python
df.head()
```

```python
df.drop(columns=['sibsp','parch','family_size'],inplace=True)
```

```python
df.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),x,y,scoring='accuracy',cv=20))
```
