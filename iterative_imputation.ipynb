```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("/content/titanic.csv",usecols=['survived','fare','age','parch'])
```

```python
from sklearn.experimental import enable_iterative_imputer
```

```python
from sklearn.impute import IterativeImputer
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.impute import SimpleImputer,IterativeImputer
#mean_imputation
mean_imputer=SimpleImputer(strategy='mean')
mean_imputed=mean_imputer.fit_transform(df)
df_mean=pd.DataFrame(mean_imputed,columns=df.columns)
```

```python
#iterative_imputation
it_imputer=IterativeImputer(random_state=42)
it_imputed=it_imputer.fit_transform(df)
df_it=pd.DataFrame(it_imputed,columns=df.columns)
```
