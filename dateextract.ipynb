```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/uni.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
#Convert date into the proper format so that date time operation can be easily performed

df["release_date"] = pd.to_datetime(df["release_date"])
df['date_year']=df['release_date'].dt.year
```

```python
df['date_year'].head()
```

```python
df.head()
```

```python
df['dt_month']=df['release_date'].dt.month
```

```python
df.head()
```

```python
df['dt_day']=df['release_date'].dt.day_name()
```

```python
df.head()
```

```python
# is weekand
df['date_is_weekand']=np.where(df['Unnamed: 2'].isin(['Sunday','Friday']),1,0)
```

```python
df.head()
```

```python
#extract week of the year
df['dt_week']=df['release_date'].dt.week
```

```python
df.head()
```

```python
#extract quarter
df['quarter']=df['release_date'].dt.quarter
```

```python
df.head()
```
